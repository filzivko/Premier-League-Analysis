```python
import pandas as pd
import requests
import json
import matplotlib.pyplot as plt
```

```python
#Total Points Plot
playerName=['Fil','Ashwin', 'Feisal',"Shaun","Angelo","Kevin", "Hayden","Darrell","Bharath","Adrian","Arzhang"]
player={'Fil':[],'Ashwin':[], 'Feisal':[],"Shaun":[],"Angelo":[],"Kevin":[], "Hayden":[],"Darrell":[],"Bharath":[],"Adrian":[],"Arzhang":[]}
playerID=['3890712','1504214','1388122','3900425','892389','1901110','4294144','5491053',"4367501","427248","4254790"]

x = []

for j in range(len(playerName)):
  url = 'https://fantasy.premierleague.com/api/entry/'+str(playerID[j])+'/history/'
  r = requests.get(url)
  json = r.json()
  elements_df = pd.DataFrame(json['current'])
  for i in range(len(elements_df)):
    player[playerName[j]].append(elements_df.total_points[i])

for l in range(len(elements_df)):
    x.append(l+1)

for k  in range(len(playerName)):
  plt.plot(x, player[playerName[k]], label=playerName[k],marker="o")

plt.title("Overall Points", fontsize=30)
plt.legend(bbox_to_anchor=(1.05, 1), loc='upper left', borderaxespad=0., fontsize=20)
plt.rcParams['figure.figsize'] = [15, len(x)*2]
plt.rcParams['figure.dpi'] = 200
plt.xlabel('Gameweek', fontsize=20)
plt.ylabel('Total Points', fontsize=20)
plt.grid()
plt.show()

print(player)
```

```python
#Per Gameweek Plot
playerName=['Fil','Ashwin', 'Feisal',"Shaun","Angelo","Kevin", "Hayden","Darrell","Bharath","Adrian","Arzhang"]
player={'Fil':[],'Ashwin':[], 'Feisal':[],"Shaun":[],"Angelo":[],"Kevin":[], "Hayden":[],"Darrell":[],"Bharath":[],"Adrian":[],"Arzhang":[]}
playerID=['3890712','1504214','1388122','3900425','892389','1901110','4294144','5491053',"4367501","427248","4254790"]

x = []



for j in range(len(playerName)):
  url = 'https://fantasy.premierleague.com/api/entry/'+str(playerID[j])+'/history/'
  r = requests.get(url)
  json = r.json()
  elements_df = pd.DataFrame(json['current'])
  for i in range(len(elements_df)):
    player[playerName[j]].append(elements_df.points[i])
  
for l in range(len(elements_df)):
    x.append(l+1)
for k  in range(len(playerName)):
  plt.plot(x, player[playerName[k]], label=playerName[k],marker="o")

plt.title("Score Per Gameweek", fontsize=30)
plt.legend(bbox_to_anchor=(1.05, 1), loc='upper left', borderaxespad=0., fontsize=20)
plt.rcParams['figure.figsize'] = [15, len(x)*2]
plt.rcParams['figure.dpi'] = 200
plt.xlabel('Gameweek', fontsize=20)
plt.ylabel('Total Points', fontsize=20)
plt.grid()
plt.show()


```
